```python
# create a guessing game where user will enter a number you have tell number is big or small
# loop user will enter another number make it untill user guess the number
# add num,subs num
num = np.random.randint(0,10)
user = int(input("enter a number :"))
while(num != user):
    user = int(input("enter a number :"))
    if (num == user):
        print("you won")
    elif(num>user):
        print("guess a large number")
    else:
        print("guess a small number ")

```
